# Notebook 02: Climate Tipping Points and Risk

In this notebook, I introduces climate tipping points
and explains how they fundamentally alter climate policy analysis.


## Units and probabilities

- Carbon stock \\(C_t\\): GtCO₂
- Tipping hazard \\(h(C)\\): probability per year (dimensionless)


## 1. Climate tipping points

A climate tipping point is a threshold beyond which
small additional changes trigger large, abrupt, and often irreversible effects.

Examples:
- Ice sheet collapse
- Amazon rainforest dieback
- Disruption of ocean circulation


## 2. Why tipping points matter for economics

Traditional models assume:
- Gradual damages
- Smooth responses

Tipping points violate this by introducing:
- Non-linearity
- Irreversibility
- Deep uncertainty


## 3. Modeling tipping points probabilistically

We do NOT assume we know exactly when tipping occurs.

Instead, we assume:
- The probability of tipping increases with carbon stock


We define a hazard rate:

\\[
h(C_t) = h_0 \exp(\gamma C_t)
\\]

Where:
- \\( h_0 \\) is baseline risk
- \\( \gamma \\) captures sensitivity to carbon stock


### Tipping Risk Visualization

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
C = np.linspace(0, 400, 200)
h0 = 0.01
gamma = 0.02

hazard = h0 * np.exp(gamma * C)

plt.figure()
plt.plot(C, hazard)
plt.xlabel("Atmospheric carbon stock (GtCO₂)")
plt.ylabel("Tipping hazard")
plt.title("Tipping risk increases with carbon stock")
plt.show()
```

### Interpretation

- Risk grows nonlinearly
- Small additional emissions at high carbon levels are very dangerous
- This creates strong incentives for precaution


## 4. Damages before and after tipping

Here, I distinguish two regimes:

Before tipping:
- Damages are low or moderate

After tipping:
- Economic output permanently drops by a fraction \\( \theta \\)


## 5. Expected welfare under tipping risk

Policy must account for:
- The chance tipping never occurs
- The chance it occurs tomorrow
- The irreversible loss if it occurs

This requires expected utility, not deterministic optimization.


## Key takeaway

Tipping points transform climate policy into:
- A risk management problem
- A precautionary decision problem
- A problem of avoiding catastrophic outcomes
