# Notebook 06: Regional Heterogeneity and Climate Policy under Tipping Risk

In this notebook, I extend the climate policy model to a multi-region world.

My objective is to:
- represent heterogeneity in emissions, damages, and responsibilities,
- study how tipping-point risk interacts with regional inequality,
- understand equity–efficiency trade-offs in global climate policy.


## Units

Throughout this notebook:

- Time step: 1 year
- Carbon stock \\( C_t \\): GtCO₂
- Regional emissions \\( E_{i,t} \\): GtCO₂/year
- Regional carbon removal \\( R_{i,t} \\): GtCO₂/year
- Welfare and value functions: dimensionless
- Regions indexed by \\( i \\)


## Why regional heterogeneity matters

Climate change is a global externality, but:
- emissions are region-specific,
- damages are unevenly distributed,
- mitigation costs differ across regions.

Ignoring heterogeneity hides the core political
and ethical challenges of climate policy.


## Regions and notation

I consider a world with \\( N \\) regions, indexed by \\( i = 1, \dots, N \\).

For each region \( i \):
- \\( E_{i,t} \\): emissions (GtCO₂/year)
- \\( R_{i,t} \\): carbon removal (GtCO₂/year)
- \\( u_i(\cdot) \\): regional welfare contribution

Atmospheric carbon is global.


## Global carbon dynamics

Atmospheric carbon accumulates according to:

\\[
C_{t+1} = C_t + \sum_{i=1}^N E_{i,t} - \sum_{i=1}^N R_{i,t}
\\]

Carbon is perfectly mixed in the atmosphere,
so regional emissions affect global climate risk.


## Regional welfare

Each region derives welfare from its own emissions and removal:

\\[
u_i(E_{i,t}, R_{i,t}) = \alpha_i \log(E_{i,t}) - \frac{\phi_i}{2} R_{i,t}^2
\\]

Where:
- \\( \alpha_i \\): weight capturing income or development level
- \\( \phi_i \\): regional cost of carbon removal


## Interpretation of heterogeneity

- High \\( \alpha_i \\): emissions are more valuable (developing regions)
- Low \\( \alpha_i \\): emissions yield less marginal welfare
- High \\( \phi_i \\): carbon removal is expensive
- Low \\( \phi_i \\): cheap removal opportunities

This structure captures global inequality
without introducing full macroeconomic models.


## Social planner objective

I consider a global social planner who maximizes
a weighted sum of regional welfare:

\\[
W_t = \sum_{i=1}^N \omega_i u_i(E_{i,t}, R_{i,t})
\\]

Where \\( \omega_i \\) are equity weights.


## Equity weights \\( \omega_i \\)

Equity weights reflect ethical choices:

- Equal weights: pure efficiency
- Higher weights on poorer regions: equity-oriented policy

These weights determine how global policy
balances efficiency and fairness.


## Global tipping-point risk

The probability of a climate tipping event depends on global carbon:

\\[
h(C_t) = h_0 \exp(\gamma C_t)
\\]

Tipping affects all regions simultaneously,
but damages may be region-specific.


## Post-tipping damages

After a tipping point, each region experiences
a permanent welfare loss \\( \Delta_i \\):

\\[
V_i^{post} = (1 - \Delta_i)\frac{u_i(\bar{E}_i)}{1 - \beta}
\\]

Damages can differ substantially across regions.


## Bellman equation with regional heterogeneity

Let \\( V(C_t) \\) denote global expected welfare.

The Bellman equation is:

\\[
\begin{aligned}
V(C_t) = \max_{\{E_{i,t}, R_{i,t}\}} \Big\{
& \sum_{i=1}^N \omega_i u_i(E_{i,t}, R_{i,t}) \\
& + \beta \Big[
(1 - h(C_t)) V(C_{t+1})
+ h(C_t) \sum_{i=1}^N \omega_i V_i^{post}
\Big]
\Big\}
\end{aligned}
\\]


## Optimality in a heterogeneous world

An optimal policy now means choosing regional emissions
and removal paths that maximize global expected welfare,
subject to physical climate constraints
and ethical equity weights.


## Efficiency versus equity

This framework highlights a core tension:

- Efficiency: minimize total global costs
- Equity: protect vulnerable regions

Different choices of \\( \omega_i \\)
generate different climate policy outcomes.


## Coordination and free-riding

Although climate risk is global:
- mitigation costs are local,
- benefits are shared.

This creates incentives for free-riding,
which explains the difficulty of international climate agreements.


## Regional carbon removal

Regions with:
- low \\( \phi_i \\) specialize in carbon removal,
- high \\( \phi_i \\) focus on mitigation.

This generates gains from international cooperation
and carbon markets.


## Interpretation for climate policy

The model suggests that:
- efficient policy allocates mitigation unevenly,
- equity concerns justify transfers or differentiated responsibilities,
- carbon pricing and carbon markets can support cooperation.

This notebook moves beyond a single global planner
and introduces the political economy of climate change.

It provides a formal framework for analyzing:
- fairness,
- burden sharing,
- international coordination,
under climate tipping risk.
