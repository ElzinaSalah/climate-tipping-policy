## Why climate clubs?

Even with carbon pricing and transfers, a key problem remains:
not all regions may participate voluntarily.

Some regions may:
- refuse to price carbon,
- free-ride on others’ mitigation,
- gain a competitiveness advantage.

Climate clubs address this by:
- rewarding participation,
- penalizing non-participation.


## Core idea of a climate club

A climate club consists of:
- members who adopt a common carbon price,
- non-members who face trade penalties.

The main enforcement tool is a
border carbon adjustment (BCA).


## Border carbon adjustment (intuition)

A border carbon adjustment:
- taxes imports from non-members,
- based on their embodied carbon,
- equalizes carbon costs across regions.

This removes the incentive to free-ride.


## Trade exposure

I introduce:
- \\( s_i \\): share of region \\( i \\)'s output that is traded
- \\( b_t \\): border tax rate (currency per GtCO₂)

For simplicity, I assume:
- Region 1 is a club member,
- Region 2 may opt out.


## Welfare with border carbon adjustment

If region \\( i \\) is outside the club, its welfare becomes:

\\[
u_i(E_{i,t}, R_{i,t})
- \\tau_t E_{i,t}
+ \\tau_t R_{i,t}
- s_i b_t E_{i,t}
+ T_{i,t}
\\]

The last term represents border penalties on exports.


## Border tax rule

I set the border tax equal to the club carbon price:

\\[
b_t = \\tau_t
\\]

This ensures:
- no competitive advantage for non-members,
- consistency with WTO principles.


## Participation decision

Region 2 compares:
- welfare inside the club,
- welfare outside the club (with border tax).

It joins the club if:

\\[
V_2^{club}(C_t) \ge V_2^{out}(C_t)
\\]


```python
# Trade exposure
s1, s2 = 0.3, 0.5  # share of output exposed to trade
```

```python
def welfare_with_club(E, R, alpha, phi, tau, transfer):
    return alpha * np.log(E) - 0.5 * phi * R**2 - tau * E + tau * R + transfer

def welfare_outside_club(E, R, alpha, phi, tau, transfer, s):
    return (
        alpha * np.log(E)
        - 0.5 * phi * R**2
        - tau * E
        + tau * R
        - s * tau * E
        + transfer
    )
```

```python
W2_club = welfare_with_club(
    policy_E2, policy_R2, alpha2, phi2, carbon_price, transfer2
)

W2_out = welfare_outside_club(
    policy_E2_nash, policy_R2_nash, alpha2, phi2, carbon_price, transfer2, s2
)
```

```python
plt.figure()
plt.plot(C_grid, W2_club, label="Inside climate club")
plt.plot(C_grid, W2_out, "--", label="Outside climate club")
plt.xlabel("Atmospheric carbon stock (GtCO₂)")
plt.ylabel("Regional welfare (dimensionless)")
plt.title("Incentives to join the climate club")
plt.legend()
plt.show()
```

## Interpretation

The border carbon adjustment:
- reduces free-riding incentives,
- makes club membership attractive,
- stabilizes cooperation.

As carbon risk increases,
joining the club becomes strictly optimal.


## Climate clubs vs transfers

- Transfers address equity among participants.
- Climate clubs address participation incentives.

Both instruments are complementary,
not substitutes.


## Real-world analogues

This framework corresponds to:

- coordinated carbon pricing agreements.

It explains why border measures are central
to future climate governance.
