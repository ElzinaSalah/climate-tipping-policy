## The core problem

Climate change is not smooth and gradual.

Instead, it may involve tipping points:
sudden, irreversible changes such as:
- ice sheet collapse,
- rainforest dieback,
- major circulation shifts.

Once crossed, these changes cannot be undone.

Policy must therefore manage risk, not just averages.


## Why standard policy falls short

Many climate policies assume:
- gradual damages,
- predictable responses,
- independent national actions.

This ignores three realities:
1. climate risk is global,
2. damages are nonlinear,
3. countries act strategically.

Ignoring these leads to under-mitigation.


## What this analysis does

I built a quantitative model that:
- tracks atmospheric carbon over time,
- links carbon levels to tipping risk,
- allows countries to act strategically,
- evaluates different policy instruments.

The model is designed to answer:
"What policies work *in the real world*?"


## Key finding 1: Delay is extremely costly

As atmospheric carbon accumulates:
- tipping risk rises rapidly,
- future policy becomes less effective,
- costs increase nonlinearly.

Early action dramatically reduces:
- long-run costs,
- catastrophic risk,
- need for extreme measures later.


## Key finding 2: Acting alone does not work

When countries act independently:
- emissions are too high,
- carbon removal is too low,
- global risk remains excessive.

This is a free-rider problem:
everyone benefits from mitigation,
but each country prefers others to act.


## Key finding 3: Carbon pricing is necessary — but not sufficient

A carbon price:
- internalizes climate damages,
- provides clear incentives,
- encourages clean investment.

However:
- pricing alone can be politically infeasible,
- distributional concerns matter,
- participation is not guaranteed.


## Key finding 4: Transfers enable cooperation

Revenue from carbon pricing can be:
- redistributed domestically,
- transferred internationally,
- used for climate finance.

Transfers:
- do not reduce efficiency,
- improve equity,
- increase political acceptability.


## Key finding 5: Climate clubs work

Climate clubs combine:
- common carbon pricing among members,
- border carbon adjustments for non-members.

This:
- removes free-riding incentives,
- protects competitiveness,
- encourages countries to join voluntarily.


## Border carbon adjustments — in simple terms

A border carbon adjustment:
- charges imports for their carbon content,
- applies the same carbon cost to domestic and foreign producers,
- prevents "carbon leakage".

This levels the playing field
and rewards participation.


## A policy package that works

The analysis supports a three-pillar strategy:

1. A rising global carbon price,
2. Targeted transfers and climate finance,
3. Climate clubs with border adjustments.

Together, these:
- reduce emissions efficiently,
- manage tipping risk,
- sustain international cooperation.


## Consequences of ignoring tipping risk

If tipping risk is ignored:
- policies will be too weak,
- carbon budgets will be exceeded,
- irreversible damages become likely.

This is not a marginal issue —
it fundamentally changes optimal policy.


## Practical implications

For governments, this means:

- carbon prices should reflect risk, not averages,
- climate finance is not charity, but efficiency-enhancing,
- border measures are a cooperation tool, not protectionism,
- early action is vastly cheaper than delay.


## What this analysis does NOT assume

This analysis does NOT assume:
- perfect global cooperation,
- altruistic governments,
- unrealistic technology breakthroughs.

It explicitly models:
- self-interest,
- strategic behavior,
- political constraints.


## Connection to real policy debates

## Final message

Climate policy is not only about reducing emissions.

It is about:
- managing catastrophic risk,
- aligning incentives,
- sustaining cooperation.

Well-designed policy can do all three.
