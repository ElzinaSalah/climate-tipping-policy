# Notebook 01: Climate–Carbon Fundamentals

This notebook introduces the physical and economic foundations
of climate–carbon dynamics.

Here, I am answering three core questions:
1. Why is carbon a stock pollutant?
2. How do emissions accumulate over time?
3. Why does this create an intertemporal policy problem?


## Units

- Time step: 1 year
- Carbon stock \\(C_t\\): GtCO₂
- Emissions \\(E_t\\): GtCO₂/year


## 1. Carbon as a Stock Pollutant

Carbon dioxide (CO₂) differs from local pollutants because it:
- Persists in the atmosphere for decades to centuries
- Accumulates over time
- Affects global temperature through cumulative emissions

This means climate change is driven by the stock of carbon, not just annual emissions.


## 2. Carbon stock dynamics

Let:
- \\( C_t \\): atmospheric carbon stock
- \\( E_t \\): emissions during period \\( t \\)

The simplest carbon accumulation equation is:

\\[
C_{t+1} = C_t + E_t
\\]

This equation expresses mass balance:
what is emitted today remains in the atmosphere tomorrow.


This equation implies:
- Past emissions constrain future choices
- Delayed mitigation makes the problem harder
- Climate policy is inherently intertemporal


### Numerical Example

```python
import numpy as np
import matplotlib.pyplot as plt

T = 50
E = np.ones(T) * 10  # constant emissions
C = np.zeros(T)

for t in range(1, T):
    C[t] = C[t-1] + E[t]

plt.figure()
plt.plot(C)
plt.xlabel("Time (years)")
plt.ylabel("Atmospheric carbon stock (GtCO₂)")
plt.title("Carbon accumulation with constant emissions")
plt.show()
```

Carbon accumulation is measured in gigatonnes of CO₂, consistent with
global carbon budget accounting used by the IPCC.


### Interpretation

- Even constant emissions lead to rising carbon
- Stabilizing emissions does NOT stabilize climate
- Carbon stock keeps growing unless emissions go to zero


## 3. From carbon stocks to carbon budgets

Climate science shows that temperature increase is approximately linear
in cumulative CO₂ emissions.

This leads to the concept of a carbon budget:
a fixed total amount of CO₂ that can be emitted
while staying below a temperature target (e.g. 1.5°C).


## Key takeaway

Climate change is a stock problem.
This creates:
- Intertemporal trade-offs
- Scarcity of emissions
- The economic logic of carbon pricing and budgets
