## Why a non-cooperative Nash equilibrium?

So far, I assumed a global social planner who coordinates emissions
and carbon removal across regions.

However, in reality:
- countries act independently,
- each region maximizes its own welfare,
- climate damages are global but mitigation costs are local.

This creates a free-rider problem.

To capture this, I now model climate policy as a non-cooperative Nash game
between two regions.


## Planner vs Nash: conceptual difference

- Social planner:
  - internalizes global climate damages,
  - coordinates emissions and removals,
  - maximizes weighted global welfare.

- Nash equilibrium:
  - each region chooses its own policy,
  - takes the other region’s actions as given,
  - ignores damages imposed on others.

This difference is the core source of under-mitigation.


## Regional optimization problem

Region \\( i \\) solves:

\\[
\max_{E_{i,t}, R_{i,t}} 
\; u_i(E_{i,t}, R_{i,t})
+ \beta \mathbb{E}_t \left[ V_i(C_{t+1}) \right]
\\]

subject to:

\\[
C_{t+1} = C_t + E_{i,t} + E_{j,t} - R_{i,t} - R_{j,t}
\\]

where \\( j \neq i \\).


## Bellman equations in Nash equilibrium

Region 1 solves:

\\[
\begin{aligned}
V_1(C_t) = \max_{E_1,R_1} \Big\{
& u_1(E_{1,t}, R_{1,t}) \\\\
& + \beta \big[
(1 - h(C_t)) V_1(C_{t+1}) + h(C_t) V_1^{post}
\big]
\Big\}
\end{aligned}
\\]

Region 2 solves an analogous problem.

A Nash equilibrium is a fixed point in policy functions:
each region’s policy is optimal given the other’s.


## Source of inefficiency

Each region ignores:
- climate damages imposed on the other region,
- the global effect of emissions on tipping risk.

As a result:
- emissions are too high,
- carbon removal is too low,
- tipping risk is inefficiently large.

This mirrors real-world climate negotiations.


```python
V1 = np.zeros(len(C_grid))
V2 = np.zeros(len(C_grid))

policy_E1_nash = np.zeros_like(V1)
policy_R1_nash = np.zeros_like(V1)

policy_E2_nash = np.zeros_like(V2)
policy_R2_nash = np.zeros_like(V2)

V1_post = -30.0
V2_post = -30.0
```

```python
for _ in range(100):
    # --- Region 1 best response ---
    for i, C in enumerate(C_grid):
        best_val = -1e12
        for E1 in E_grid:
            for R1 in R_grid:
                # take region 2 policy as given
                E2 = policy_E2_nash[i]
                R2 = policy_R2_nash[i]

                C_next = C + E1 + E2 - R1 - R2
                if C_next < C_grid[0] or C_next > C_grid[-1]:
                    continue

                j = np.searchsorted(C_grid, C_next)
                h = hazard(C)

                val = (
                    u_region(E1, R1, alpha1, phi1)
                    + beta * ((1 - h) * V1[j] + h * V1_post)
                )

                if val > best_val:
                    best_val = val
                    V1[i] = val
                    policy_E1_nash[i] = E1
                    policy_R1_nash[i] = R1

    # --- Region 2 best response ---
    for i, C in enumerate(C_grid):
        best_val = -1e12
        for E2 in E_grid:
            for R2 in R_grid:
                E1 = policy_E1_nash[i]
                R1 = policy_R1_nash[i]

                C_next = C + E1 + E2 - R1 - R2
                if C_next < C_grid[0] or C_next > C_grid[-1]:
                    continue

                j = np.searchsorted(C_grid, C_next)
                h = hazard(C)

                val = (
                    u_region(E2, R2, alpha2, phi2)
                    + beta * ((1 - h) * V2[j] + h * V2_post)
                )

                if val > best_val:
                    best_val = val
                    V2[i] = val
                    policy_E2_nash[i] = E2
                    policy_R2_nash[i] = R2
```

```python
plt.figure()
plt.plot(C_grid, policy_E1, label="Planner: Region 1")
plt.plot(C_grid, policy_E1_nash, "--", label="Nash: Region 1")
plt.plot(C_grid, policy_E2, label="Planner: Region 2")
plt.plot(C_grid, policy_E2_nash, "--", label="Nash: Region 2")

plt.xlabel("Atmospheric carbon stock (GtCO₂)")
plt.ylabel("Emissions (GtCO₂/year)")
plt.title("Planner vs Nash emissions")
plt.legend()
plt.show()
```

```python
plt.figure()
plt.plot(C_grid, policy_R1, label="Planner: Region 1")
plt.plot(C_grid, policy_R1_nash, "--", label="Nash: Region 1")
plt.plot(C_grid, policy_R2, label="Planner: Region 2")
plt.plot(C_grid, policy_R2_nash, "--", label="Nash: Region 2")

plt.xlabel("Atmospheric carbon stock (GtCO₂)")
plt.ylabel("Carbon removal (GtCO₂/year)")
plt.title("Planner vs Nash carbon removal")
plt.legend()
plt.show()
```

## Interpretation of Nash equilibrium results

Compared to the social planner:

- emissions are higher under Nash behavior,
- carbon removal is systematically lower,
- tipping risk increases,
- coordination failure dominates efficiency.

This result formalizes the climate free-rider problem
in a dynamic tipping-risk framework.


## Real-world interpretation

The Nash equilibrium corresponds to:
- nationally determined contributions (NDCs),
- fragmented climate policies,
- weak international enforcement.

The planner solution represents:
- globally coordinated carbon pricing,
- international transfers,
- shared responsibility mechanisms.
